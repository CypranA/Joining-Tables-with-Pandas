```python
movies = pd.read_pickle('movies.p')
movies
```

```python
taglines = pd.read_pickle('taglines.p')
taglines
```

```python
#merge with left join
#returns all the rows of the table on the left side of the join and matching rows for the table on the right side of join. The rows for which there is no matching row on right side, the result-set will contain null. 
#when there isnt a match, the columns on the right table will be missing

movie_taglines = movies.merge(taglines, on='id', how='left')
movie_taglines
```

```python
#counting the number of missing rows in a dataframe or table
#isnull funtion is used to check the number of rows that are missing in a dataframe

number_of_missing_tagline = movie_taglines['tagline'].isnull().sum()
number_of_missing_tagline
```

```python
#merging with right join
#Right joins are similar to left joins except they return all rows from the table in the RIGHT JOIN clause and only matching rows from the table in the FROM clause
#when there isnt a match, the columns on the left table will be missing

movie_to_genre = pd.read_pickle('movie_to_genres.p')
#filtering for TV Movie genre
tv_genre = movie_to_genre[movie_to_genre['genre'] =='TV Movie']
tv_genre
```

```python
movies
```

```python
#merging movies and tv_genre
#movies table has a column name 'id' whilst tv_genre table have a column name 'movie_id'. Both are the same so we use the left_on and right_on function to make them the same.
tv_movies = movies.merge(tv_genre, how='right', left_on='id', right_on='movie_id')
tv_movies
```

```python
#merging with outer join
#outer join merges both or more dataframes together regardless of whether there is a match or not. ALl rows from the tables are displayed

family = movie_to_genre[movie_to_genre['genre']=='Family']
comedy = movie_to_genre[movie_to_genre['genre']=='Comedy']
print(family.head())
print(comedy.head())
```

```python
#outer join
family_comedy = family.merge(comedy, on='movie_id', how='outer', suffixes=['_fam', '_com'])
family_comedy
```

# verifying integrity using merge method

verifying integrity is system of checking if there are errors with the dataframes you want to merge or work on.
it helps avoid duplication of values

```python
#verifying integrity using the validate argument
#this helps to avoid duplication of values/data when multiple dataframes/tables are merged
```

```python
tracks = pd.read_csv('tracks.csv')
tracks
```

```python
specs = pd.read_csv('specs.csv')
specs
```

```python
tracks.merge(specs, on='tid')
```

```python
tracks.merge(specs, on='tid', validate='one_to_one')

#validating a one to one relationship for the merged dataframe (track and specs dataframes) shows an error since the specs dataframe has two rows with similar values 
```

```python
tracks.merge(specs, on='tid', validate='one_to_many')

#validating a one to many relationship shows an executable result since specs has 2 rows with similar value
```

# merge_ordered()

. just like the merge method, the merge_ordered() is used to sort the rows of the merged dataframe
. the default merge without indicating the 'how' is 'outer' (how='outer') whilst the default join for merge is 'inner'
. pandas (pd) has to be called before the merge can be done
. the code for merge_ordered() is: pd.merge_ordered(df1, df2)

```python
appl = pd.read_csv('appl.csv')
appl
```

```python
mcd = pd.read_csv('mcd.csv')
mcd
```

```python
pd.merge_ordered(appl, mcd, on='date', suffixes=['_appl', '_mcd'])

#the join method ('how') is already set at 'outer' by default
#the values in the dataframe are ordered in ascending order due to the dates 
```

```python
pd.merge_ordered(appl, mcd, on='date', how='left', suffixes=('_appl', '_mcd'))
```

```python
#foward fill ('ffill')
#filling the missing data with the value of the previous data

pd.merge_ordered(appl, mcd, on='date', suffixes=('_appl', '_mcd'), fill_method='ffill')

#the missing data in the first close column is not filled since there is no previous data available to fill it in
```

# merge_asof()

. similar to merge_ordered
. This is similar to a left-join except that we match on nearest key rather than equal keys. Both DataFrames must be sorted by the key.

```python
visa = pd.read_csv('visa.csv')
visa
```

```python
ibm = pd.read_csv('ibm.csv')
ibm
```

```python
pd.merge_asof(visa, ibm, on='time', direction='forward')

#direction is used to select table on the right as the determinant of which time is nearest to match
```

# Reshaping data with .melt() method

data can come in the form of a wide format or a long format. we can manipulate the structure/shape of the data using the .melt() method

```python
social_fin = pd.read_csv('social_fin.csv')
social_fin

#we can reshape the dataframe as you can see that the years are horizontal
```

```python
#tall shape of social_fin
#id_vars stands for identifier variables. it is used to select those columns we dont want to change
social_fin_tall = social_fin.melt(id_vars=['financial', 'company'])
social_fin_tall

#the result transforms the years into one column
#two new columns are generated. variable to represent the years and value to represent the figures for each year
```

```python
#value_vars is used to show which values are unpivoted
#
social_fin_tall = social_fin.melt(id_vars=['financial', 'company'], value_vars=['2018', '2017'])
social_fin_tall
```

```python
#meltfing with column names
#changing the names of the new columns after melting
#using the var_name and vlaue_name argument
social_fin_tall = social_fin.melt(id_vars=['financial', 'company'], value_vars=['2018', '2017'], var_name='year', value_name='dollars')
social_fin_tall
```
